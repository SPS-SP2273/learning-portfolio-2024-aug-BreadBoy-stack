```python
from matplotlib import pyplot as plt
import numpy as np
```

# What to expect in this chapter

```python
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)
```

# 1 Some nomenclature

```python
plt.plot(x, cos_x, label='cos x')
plt.plot(x, sin_x, label='sin x')
plt.legend()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax[0].plot(x, cos_x, label='cos x')
ax[1].plot(x, sin_x, label='sin x')
ax.legend()

ax[0].plot(0)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(10,5),sharex=True, sharey=True)

for count, a in enumerate(ax.flatten()):
    if count == 3:
        a.plot(x, 10*cos_x)
        continue
        
    a.plot(x, sin_x)
    a.grid(alpha=0.25)
    a.spines['right'].set_visible(False)
    a.spines['left'].set_visible(False)
    a.spines['top'].set_visible(False)
```

```python
ax[1,1].plot(x, cos_x)
ax[1,2].plot(cos_x, sin_x)
```

# 4 A complete OO example

```python
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(5, 5),
                       sharex='col', sharey='col')

# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)

x_2 = np.arange(10)

bottom_left.bar(x_2, x_2**2)
bottom_right.plot(x, sin_x)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
```
